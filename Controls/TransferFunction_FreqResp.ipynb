```julia
#Robot joint driven by DC motor through a gear transmission w/ ratio r:1
#θₘ(s)/V(s) = (kt/R) / (s*(Jₘs + B + (kₑkₜ/R)))
#θₘ(s)/τₗ(s) = (-1 / r) / (s * (Jₘs + Bₘ + (kₑkₜ/R)))

#Cummulatively,
# J θ_dot_dot(t) + B θ_dot(t) = u(t) - d(t)
# where,   B = Bₘ + kₑkₜ/R
#          u = (kₜ/R)V
#          d = τₗ/r

# Final TF: P(s) = 1 / (s * (Js + B))
```

```julia
#CT transfer function
P₁ = tf([1],[10, 0.1, 0])
```

```julia
begin
    s = tf("s")
    P₂ = 1/(10*s^2 + 0.1*s)
end
```

```julia
#Frequency Response
# (jω) = P(s)|(s = jω)
```

```julia
begin
    setPlotScale("dB")
    fig = bodeplot(P₂; title="Bode plot of P(s)", label="\$P_2\$")
end
```

```julia
#Assume Joint angle is measured by potentiometer and resulting voltage is fed into 12-bit A/D of microcontroller
#0⁰ - 360⁰ => 0 - 4095 (4095/ 360 = 11.37 scaling factor for the TF)
#New TF:
#      P(s) = 11.375 / (10s² + 0.1s)
#Bandwidth: 1 rad/s
```

```julia
begin
    num3 = 11.375
    den3 = [10, 0.1, 0]
    P₃ = tf(num3, den3)
    w = exp10.(LinRange(-2,2,2000))
    #w = logrange(0.01, 100, 200)
    bodeplot(P₃, w; title="Bode Plot of P(s)", label="\$P_3\$")
end
```

```julia
#Evaluate Bandwidth
begin
    mag, phase, w1 = bode(P₃, w)
    ivec = Tuple.(findall(x->x<0.707, mag))
    ivec
    index = last.(ivec)[1]
    w1[index]
end
```

```julia
#Assume output is corrupted with noise: sine wave of mag 0.5 and freq 5Hz
#This can be gotten rid of by using a Passive LPF w/ cutoff freq at plant bandwidth
#LPF : R = 60KΩ C = 10μF
# H(s) = 1 / (RCs + 1)
```
